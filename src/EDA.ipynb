# Análise entre as variáveis categóricas

```python
#  Filtragem dos dados categóricos apenas
categorical_vars = [col for col in dataset_features.columns if col not in ["age", "diameter_1", "diameter_2", "fitspatrick"]]
categorical_vars
```

```python
from scipy.stats import chi2_contingency


for var in categorical_vars:
    plt.figure(figsize=(12, 6))
    sns.countplot(data=dataset_features, x='diagnostic', hue=var, palette='Set3')
    plt.title(f'Relação entre {var} e Diagnóstico')
    plt.xlabel('Diagnóstico')
    plt.ylabel('Contagem')
    plt.legend(title=var)
    plt.show()

    # Teste Qui-Quadrado
    contingency_table = pd.crosstab(dataset_features[var], dataset_features['diagnostic'])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    print(f"Variável: {var}")
    print(f"Qui-quadrado: {chi2}, p-valor: {p}")
    if p < 0.05: # Para valores abaixo de 5% é considerado a hipótese de que há uma relação significativa
        print(f"Há uma relação significativa entre {var} e diagnostic.\n")
    else:
        print(f"Não há uma relação significativa entre {var} e diagnostic.\n")
```

Verificar a relação entre as variáveis categóricas de forma individual

```python
columns_original = list(dataset_features.columns)
columns_numericas = ['age', 'fitspatrick', 'diameter_1', 'diameter_2']
columns_categoricas = list(set(columns_original) - set(columns_numericas))
print("Colunas categóricas:", columns_categoricas)
```

Verificação da co-ocorência

```python
# Loop sobre cada variável categórica em 'columns_categoricas'
for categorical_category in columns_categoricas:
    # Calcula a tabela cruzada normalizada entre a variável categórica e 'diagnostic'
    results = pd.crosstab(dataset_features[categorical_category], 
                          dataset_features['diagnostic'], 
                          normalize='index')
    
    # Exibe o nome da variável categórica e os resultados da tabela cruzada
    print(f"Tabela cruzada para '{categorical_category}' em relação a 'diagnostic':\n")
    print(results)
    print("\n" + "-"*50 + "\n")
```

Verificação do valor chi² e p, para saber a relação de uma com outra

```python
from scipy.stats import chi2_contingency

for categorical_category in columns_categoricas:
    # Calcula o teste qui-quadrado
    print(f"Para a análise da variável categórica: {categorical_category}\n")
    tabela_contingencia = pd.crosstab(dataset_features[categorical_category], dataset_features['diagnostic'])
    chi2, p, _, _ = chi2_contingency(tabela_contingencia)
    print(f'Qui-quadrado: {chi2}, Valor-p: {p}')
```

Análise de componentes múltiplas

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Codificar as variáveis categóricas
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(dataset_features[categorical_vars]).toarray()

# Converter para DataFrame com os nomes das colunas
encoded_features = encoder.get_feature_names_out(categorical_vars)
encoded_df = pd.DataFrame(encoded_data, columns=encoded_features)
```

```python
import prince

# Instanciar a MCA
mca = prince.MCA(n_components=60, random_state=42, n_iter=100)

# Ajustar e transformar os dados
mca_result = mca.fit(encoded_df)
```

```python
# Obtenção dos alto valores dos componentes
mca_result.eigenvalues_summary
```

```python
# Obter as coordenadas das linhas
mca_df = mca.row_coordinates(encoded_df)
mca_df
```

Análise dos alto valores por componente

```python
eigenvalues = mca.eigenvalues_
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, marker='o')
plt.title('Variância dos alto-valores por componente')
plt.xlabel('Número de Componentes')
plt.ylabel('Valor Próprio (Eigenvalue)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Obter as coordenadas das linhas no espaço dos componentes principais
row_coords = mca.row_coordinates(encoded_df)

# Obter as coordenadas das colunas (variáveis)
column_coords = mca.column_coordinates(encoded_df)

import matplotlib.pyplot as plt

# Criar o biplot
plt.figure(figsize=(12, 8))

# Plotar as coordenadas das linhas (indivíduos)
plt.scatter(row_coords[0], row_coords[1], alpha=0.5, label='Indivíduos')

# Plotar as coordenadas das colunas (categorias das variáveis)
for i, (x, y) in enumerate(zip(column_coords[0], column_coords[1])):
    # Evitar que o índice ultrapasse o número de colunas
    if i < len(encoded_df.columns):
        plt.scatter(x, y, color='red', label='Categorias' if i == 0 else "")
        plt.text(x, y, encoded_df.columns[i], fontsize=10, color='red')

# Customizar o gráfico
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.title('Biplot das Categorias e Indivíduos')
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
